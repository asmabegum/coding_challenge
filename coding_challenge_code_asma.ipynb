```python
#Libraries
import pandas as pd
import numpy as np
```

```python
#Importing data
customer= pd.read_csv("customer_test.csv")
order=pd.read_csv("orders_test.csv")
store=pd.read_csv("store_test.csv")
```

```python
#creating dataframe with count of orders y customer
temp_df = pd.DataFrame(order['customer_id'].value_counts())
temp_df = temp_df.T
temp_df
```

```python
# substituting the count of assignment group in the data
order['Count_cust'] = order.apply(lambda row: temp_df[row['customer_id']] , axis=1)
```

```python
#Merging customer, order and store details
cust_ord_merge = pd.merge(customer, order, left_on='id', right_on='id')
cust_ord_merge=pd.merge(cust_ord_merge,store,left_on='store_id',right_on='id')
```

```python
cust_ord_merge.head()
```

```python
#Filtering orders less than 10
cust_df=cust_ord_merge[cust_ord_merge['Count_cust']<10]
cust_df
```

```python
#creating a dataframe
final_df=pd.DataFrame()
final_df['First name']=cust_df['first_name']
final_df['Last name']=cust_df['last_name']
final_df['Email']=cust_df['email']
final_df['Orders Placed by User']=cust_df['total'] #Assuming orders placed by user is total amount
```

```python
#Encrypt using MD5
df_copy=final_df.copy()
import hashlib
val=[]
for i in df_copy['Email']:
    val.append(hashlib.md5(i.encode('utf-8')))
```

```python
val
```

```python
df_copy['Email']=val
```

```python
#Exporting the file
df1=df_copy.to_csv('csv_2.csv', index=False)
```

```python
#funtion for month variable
def mapper(month):
    return month.strftime('%B')
```

```python
# creating year, month variable
cust_ord_merge['order_date'] = pd.to_datetime(cust_ord_merge['order_date'])
cust_ord_merge['year'] = cust_ord_merge['order_date'].dt.year
#cust_ord_merge['month'] = cust_ord_merge['order_date'].dt.month
cust_ord_merge['month']=cust_ord_merge['order_date'].apply(mapper)
```

```python
cust_ord_merge.head()
```

```python
#aggregating using group by
grp_df=cust_ord_merge.groupby(['year','month','name']).agg({'id_x':'count','total':'sum'}).reset_index().rename(columns={'id_x':'Number of Orders','total':'Total Revenue'})
grp_df
```

```python
#Exporting the file
df=grp_df.to_csv('csv_1.csv', index=False)
```
